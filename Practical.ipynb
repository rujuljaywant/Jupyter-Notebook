```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = sns.load_dataset('iris')
data.head()
```

```python
data.tail()
```

```python
data.describe()
```

```python
data.columns
```

```python
data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].plot(kind='box', subplots=True)
plt.show()
```

```python
data.species.value_counts(normalize=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
data.shape
```

```python
x,xt,y,yt=train_test_split(data.drop(columns='species'),
                           data.species,
                          stratify=data.species,
                          test_size=0.1,
                          random_state=42)
```

### Bagging

### Random forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x,y)
```

```python
rf.score( xt, yt)
```

```python
rf.feature_importances_
```

```python
pd.Series(rf.feature_importances_,index=x.columns).nlargest(10).plot(kind='barh')
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
```

```python
dt=DecisionTreeClassifier()
nb=GaussianNB()
svc=SVC()
knn=KNeighborsClassifier()
lr=LogisticRegression()
```

```python
model=[("Decision Tree",dt),("Naive_Bayes",nb),("Logistic Regression",lr),("K_Nearest",knn),("Supprt_Vector",svc)]
```

```python
vc=VotingClassifier(model)
```

```python
vc.fit(x,y)
```

```python
vc.score(xt,yt)
```

```python
vc.estimators_
```
