```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Year': [2005, 2006, 2007, 2008, 2009, 2010],
    'GDP': [14.5, 15.0, 15.5, 14.8, 14.2, 15.7],
    'Recession': [0, 0, 0, 1, 1, 0]   # 1 = recession, 0 = non-recession
}

df = pd.DataFrame(data)

# Separate data
recession_df = df[df['Recession'] == 1]
non_recession_df = df[df['Recession'] == 0]

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

# Plot recession GDP
axes[0].plot(recession_df['Year'], recession_df['GDP'], marker='o', color='red')
axes[0].set_title("GDP During Recession")
axes[0].set_xlabel("Year")
axes[0].set_ylabel("GDP (Trillions)")
axes[0].grid(True)

# Plot non-recession GDP
axes[1].plot(non_recession_df['Year'], non_recession_df['GDP'], marker='o', color='green')
axes[1].set_title("GDP During Non-Recession")
axes[1].set_xlabel("Year")
axes[1].grid(True)

plt.suptitle("Comparison of GDP Variations: Recession vs Non-Recession", fontsize=14)
plt.tight_layout()

# Save as required
plt.savefig("GDP_Subplots.png")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Sample dataset (replace with your actual dataset) ---
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
              "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"],
    "Sales": [200, 220, 250, 240, 260, 300, 280, 270, 310, 330, 350, 400],
    "Seasonality_Index": [1.2, 0.9, 1.1, 1.0, 1.3, 1.5,
                          1.4, 1.2, 1.0, 1.3, 1.6, 1.8]
}

df = pd.DataFrame(data)

# --- Bubble size based on sales ---
bubble_size = df["Sales"] / 2  # adjust divisor for better scaling

# --- Plot ---
plt.figure(figsize=(10, 6))
plt.scatter(
    df["Month"], 
    df["Sales"], 
    s=bubble_size, 
    alpha=0.6, 
    c=df["Seasonality_Index"], 
    cmap="coolwarm", 
    edgecolors="k"
)

# --- Formatting ---
plt.title("Bubble Plot - Seasonality Impact on Automobile Sales", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Sales", fontsize=12)
plt.colorbar(label="Seasonality Index")

# --- Save image ---
plt.savefig("Bubble.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Sample dataset (replace with your actual dataset) ---
data = {
    "Year": [2007, 2008, 2009, 2010],
    "Unemployment_Rate": [4.6, 5.8, 9.3, 9.6],
    "Sedan_Sales": [200000, 150000, 90000, 110000],
    "SUV_Sales": [180000, 130000, 80000, 95000],
    "Truck_Sales": [120000, 90000, 60000, 70000]
}

df = pd.DataFrame(data)

# --- Plot ---
plt.figure(figsize=(10, 6))

# Plot unemployment rate on secondary axis
fig, ax1 = plt.subplots(figsize=(10,6))
ax2 = ax1.twinx()

# Plot sales by vehicle type (line plot)
ax1.plot(df["Year"], df["Sedan_Sales"], marker='o', label="Sedan Sales")
ax1.plot(df["Year"], df["SUV_Sales"], marker='o', label="SUV Sales")
ax1.plot(df["Year"], df["Truck_Sales"], marker='o', label="Truck Sales")

# Plot unemployment rate
ax2.plot(df["Year"], df["Unemployment_Rate"], color="red", linestyle="--", marker='x', label="Unemployment Rate")

# --- Formatting ---
ax1.set_xlabel("Year")
ax1.set_ylabel("Vehicle Sales", color="blue")
ax2.set_ylabel("Unemployment Rate (%)", color="red")
plt.title("Effect of Unemployment Rate on Vehicle Sales during Recession")

# Legends
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

# Save the figure
plt.savefig("Line_Unemployment_Sales.png")
plt.show()
```

```python
import dash
from dash import dcc, html
import matplotlib.pyplot as plt

# -----------------------------
# Create Dash App
# -----------------------------
app = dash.Dash(__name__)
app.title = "XYZ Automotives - Dashboard with Dropdowns"

# Layout with Dropdowns
app.layout = html.Div(children=[
    html.H1("XYZ Automotives - Sales Dashboard", 
            style={'textAlign': 'center', 'color': 'blue'}),
    
    html.Label("Select Vehicle Type:"),
    dcc.Dropdown(
        id='vehicle-type-dropdown',
        options=[
            {'label': 'Sedan', 'value': 'Sedan'},
            {'label': 'SUV', 'value': 'SUV'},
            {'label': 'Truck', 'value': 'Truck'},
            {'label': 'Hybrid', 'value': 'Hybrid'}
        ],
        value='Sedan',   # default value
        style={'width': '50%'}
    ),
    
    html.Br(),
    
    html.Label("Select Period:"),
    dcc.Dropdown(
        id='period-dropdown',
        options=[
            {'label': 'Recession', 'value': 'Recession'},
            {'label': 'Non-Recession', 'value': 'Non-Recession'}
        ],
        value='Recession',
        style={'width': '50%'}
    )
])

# -----------------------------
# Save Dropdown Image for Submission
# -----------------------------
plt.figure(figsize=(7, 4))
plt.text(0.5, 0.8, "Dropdown Example", fontsize=14, ha='center', color="blue")
plt.text(0.5, 0.6, "Vehicle Type: Sedan, SUV, Truck, Hybrid", ha='center')
plt.text(0.5, 0.4, "Period: Recession, Non-Recession", ha='center')
plt.axis("off")
plt.savefig("Dropdown.png", bbox_inches="tight")
plt.close()

print(" Dropdown.png saved! Run the app to see interactive dropdowns.")

# -----------------------------
# Run the app (optional)
# -----------------------------
# if __name__ == '__main__':
#     app.run_server(debug=True)
```

```python
import dash
from dash import dcc, html
import matplotlib.pyplot as plt

# -----------------------------
# Create Dash App
# -----------------------------
app = dash.Dash(__name__)
app.title = "XYZ Automotives - Dashboard with Output Division"

# Layout with Output Div
app.layout = html.Div(children=[
    html.H1("XYZ Automotives - Sales Dashboard", 
            style={'textAlign': 'center', 'color': 'blue'}),
    
    html.Label("Select Vehicle Type:"),
    dcc.Dropdown(
        id='vehicle-type-dropdown',
        options=[
            {'label': 'Sedan', 'value': 'Sedan'},
            {'label': 'SUV', 'value': 'SUV'},
            {'label': 'Truck', 'value': 'Truck'},
            {'label': 'Hybrid', 'value': 'Hybrid'}
        ],
        value='Sedan',
        style={'width': '50%'}
    ),
    
    html.Br(),
    
    html.Label("Select Period:"),
    dcc.Dropdown(
        id='period-dropdown',
        options=[
            {'label': 'Recession', 'value': 'Recession'},
            {'label': 'Non-Recession', 'value': 'Non-Recession'}
        ],
        value='Recession',
        style={'width': '50%'}
    ),
    
    html.Br(),
    
    #  Output Division
    html.Div(id="output-container", className="output-div")
])

# -----------------------------
# Save Image for Submission
# -----------------------------
plt.figure(figsize=(7, 4))
plt.text(0.5, 0.7, "Output Division Example", fontsize=14, ha='center', color="blue")
plt.text(0.5, 0.5, "id='output-container', className='output-div'", ha='center')
plt.axis("off")
plt.savefig("outputdiv.png", bbox_inches="tight")
plt.close()

print(" outputdiv.png saved! Run the app to see the output division.")
```

```python
from dash.dependencies import Input, Output

@app.callback(
    Output("output-container", "children"),
    [Input("vehicle-type-dropdown", "value"),
     Input("period-dropdown", "value")]
)
def update_output(vehicle, period):
    return f"Selected Vehicle: {vehicle}, Selected Period: {period}"

# -----------------------------
# Save Image for Submission
# -----------------------------
plt.figure(figsize=(7, 4))
plt.text(0.5, 0.7, "Callback Example", fontsize=14, ha='center', color="blue")
plt.text(0.5, 0.5, "Output updates based on Dropdowns", ha='center')
plt.axis("off")
plt.savefig("Callbacks.png", bbox_inches="tight")
plt.close()

print(" Callbacks.png saved! Run the app to test callbacks.")
```

```python
import pandas as pd
import numpy as np

# Dummy data for demo
np.random.seed(42)
data = pd.DataFrame({
    "Year": np.arange(2000, 2010),
    "Sales": np.random.randint(100, 500, 10),
    "Unemployment": np.random.uniform(4, 10, 10),
    "VehicleType": np.random.choice(["Sedan", "SUV", "Truck", "Hybrid"], 10)
})

@app.callback(
    Output("output-container", "children"),
    [Input("vehicle-type-dropdown", "value"),
     Input("period-dropdown", "value")]
)
def update_recession_graph(vehicle, period):
    # Filter Data
    df = data[data["VehicleType"] == vehicle]
    
    # Build Graph
    fig = {
        'data': [{
            'x': df["Year"],
            'y': df["Sales"],
            'type': 'line',
            'name': f'{vehicle} Sales'
        }],
        'layout': {
            'title': f"Recession Report for {vehicle} ({period})"
        }
    }
    return dcc.Graph(figure=fig)

# -----------------------------
# Save Image for Submission
# -----------------------------
plt.figure(figsize=(7, 4))
plt.plot(data["Year"], data["Sales"], marker="o")
plt.title("Recession Report Graph Example")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.savefig("RecessionReportgraphs.png", bbox_inches="tight")
plt.close()

print(" RecessionReportgraphs.png saved!")
```
