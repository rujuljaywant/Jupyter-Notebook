```python
import pandas as pd

data = pd.read_csv('diabetes_PCA.csv')
data.head()
```

```python
x = data.drop( columns=['class'])
y = data['class'].apply( lambda x: 1 if x == 'tested_positive' else 0)
```

###Split the dataset
The dataset is split into training (60%) and testing (40%) sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( x, y, test_size=0.4, random_state=42)
```

### Standardise the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform( X_train)
X_test_scaled = scaler.transform(X_test)
```

### Apply the PCA

- Import the PCA from sklearn.decomposition
- Find the optimal number of principal components
- Instantiate the PCA object with the optimal number of components
- Fit the PCA to the scaled data
- Transform the scaled data using PCA

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt 
import numpy as np

pca = PCA()
pca.fit(X_train_scaled)

explained_variance = pca.explained_variance_ratio_

plt.figure(figsize=(8, 4))
plt.bar(range(1, len(explained_variance) + 1), explained_variance, alpha=0.5, align='center', label='Individual explained variance')
plt.step(range(1, len(explained_variance) + 1), np.cumsum(explained_variance), where='mid', label='Cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.title('Explained Variance by Different Principal Components')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
eigenvalues = pca.explained_variance_

plt.figure(figsize=(8, 4))
plt.plot(range(1, len(eigenvalues) + 1), eigenvalues, marker='o', linestyle='-', label='Eigenvalues')
plt.xlabel('Principal components')
plt.ylabel('Eigenvalues')
plt.title('Scree Plot')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
pca = PCA (n_components=3)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

### Create dataframe

```python
df_train_pca = pd.DataFrame(data=X_train_pca, columns=['Principal component 1', 'Principal Component 2', 'Principal Component 3'])
df_train_pca['Target'] = y_train

df_train_pca = pd.DataFrame(data=X_train_pca, columns=['Principal component 1', 'Principal Component 2', 'Principal Component 3'])
df_train_pca['Target'] = y_test
```

### Visualize the results for training and testing set

```python
from mpl_toolkits.mplot3d import Axes3D  # Only needed in older matplotlib versions

fig = plt.figure(figsize=(12, 8))

# Create 3D subplot
ax = fig.add_subplot(121, projection='3d')

# Define colors for each target class
colors = ['r', 'b']
labels = ['No Diabetes', 'Diabetes']

# Plot each target group
for target, color, label in zip([0, 1], colors, labels):
    indices = df_train_pca['Target'] == target
    ax.scatter(
        df_train_pca.loc[indices, 'Principal component 1'],
        df_train_pca.loc[indices, 'Principal Component 2'],
        df_train_pca.loc[indices, 'Principal Component 3'],
        c=color,
        s=50,
        label=label
    )

# Set labels and title
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('PCA of Diabetes Dataset (Training Set)')
ax.legend()
ax.grid(True)

plt.show()
```
